```python
transactions = []
for i in range(len(df)): 
    temp = []
    for j in range(20):
        if(str(df.values[i,j]) != '0'): 
            temp.append(str(df.values[i,j]))
    transactions.append(temp)
# print(transactions)
```

```python
from apyori import apriori

rules = apriori(transactions = transactions, min_support=0.003, min_confidence=0.2, min_lift=3,min_length=2, max_length=2)
results = list(rules)
results
```

```python
def inspect(results): 
    lhs = [list(i.ordered_statistics[0].items_base)[0] for i in results]
    rhs = [list(i.ordered_statistics[0].items_add)[0] for i in results]
    support = [i.support for i in results]
    confidence = [i.ordered_statistics[0].confidence for i in results]
    lift = [i.ordered_statistics[0].lift for i in results]
    return list(zip(lhs, rhs, support, confidence, lift))

data = pd.DataFrame(inspect(results), columns=[
    'Item1', 'Item2', 'Support', 'Confidence', 'Lift'
])
data
```
